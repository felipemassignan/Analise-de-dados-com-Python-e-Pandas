```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_excel("/content/drive/MyDrive/Curso_Python_Pandas/datasets/AdventureWorks.xlsx")
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Valor Venda'].sum()
```

```python
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head()
```

```python
round(df['Custo'].sum(),2)
```

```python
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df.head()
```

```python
round(df['Lucro'].sum(),2)
```

```python
df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head()
```

```python
df['Tempo Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head()
```

```python
df['Tempo Envio'].dtypes
```

```python
df.groupby('Marca')['Tempo Envio'].mean()
```

```python
df.isnull().sum()
```

```python
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
```

```python
lucro_ano
```

```python
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='total de produtos vendidos')
plt.xlabel('Total')
plt.ylabel('Produto')
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().sort_values(ascending=True).plot.bar(title='total de produtos vendidos')
plt.xlabel('Ano')
plt.ylabel('Receita')
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().sort_values(ascending=True)
```

```python
df_2009 = df[df['Data Venda'].dt.year==2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Receita')
```

```python
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal')
```

```python
df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal')
```
